```python
import pandas as pd
```

```python
# load dataset into pandas dataframe
data = pd.read_csv("raw_songs.csv")
data
```

```python
# drop popularity column - not necessary
data = data.drop(['Popularity'], axis = 1)
data
```

```python
data["Lyrics"][40]
```

```python
count = 0
nums = []
for i in range(len(data["Lyrics"])):
    if i != 74:
        if data['Name'][i].casefold().split("-")[0] not in data["Lyrics"][i].casefold():
            print(i, data['Name'][i])
            nums.append(i)
            count += 1
print(count)
```

```python
for each in nums:
    data = data.drop(each)
data.shape
```

```python
data = data.drop(0)
```
